# 1 完整群环的构建

## 1.1 定义

对于一个投影有限群 $G$ 和一个交换环 $R$
$$
R⟦ G ⟧ := \varprojlim_{N.\mathsf{OpenNormal}}
 R[G/N]    
$$
其中逆极限作为满足坐标之间转换映射的乘积空间的子类型实现.

当 $N_1 \leqslant N_2$ 时, 转移映射 $R[G/N_1] \rightarrow R[G/N_2]$ 由典范映射 $G/N_1 \rightarrow G/N_2$ 诱导.

## 1.2 性质

### 1.2.2 
对于 $\mathsf{profinite}$ 群 $G$,若 $G = \varprojlim_{\lambda \in \Lambda} G_\lambda$,则有:
$$
R⟦ G ⟧ \cong \varprojlim_{\lambda \in \Lambda} R[G_\lambda]    
$$
特别地,当 $\Lambda = \mathbb{N}$ 时也成立.

# 2. 同构

在本节中,$R$ 表示一个 p-adic 可分完备的交换环(即 $R = \varprojlim R/(p^nR)$).

## 2.1 小同构

### 2.1.1 
若将 $[1 \bmod p^i]$ 记作 $1 + T$,则存在同构
$$
R[\mathbb{Z}/p^i\mathbb{Z}] \cong R[T] / ((1 + T)^{p^i} - 1).
$$
该同构由 $[1 \bmod p^i] \mapsto 1 + T$ 生成.

### 2.1.2 
对上述同构的两边取逆极限.

## 2.2 
$$
\varprojlim_{i} R[T] / ((1 + T)^{p^i} - 1) \cong R⟦ T ⟧    
$$
